```python
from fastapi import FastAPI
from pydantic import BaseModel
import pickle
import numpy as np

# Load the model
with open("surveillance_model.pkl", "rb") as f:
    model = pickle.load(f)

# Initialize FastAPI app
app = FastAPI()

# Define input schema
class InputData(BaseModel):
    person_id:int
    call_count_with_suspects:int
    night_movements_hours:float
    gps_zone_visits:int
    encrypted_app_usage_hours:float
    social_media_threat_score:float
    vpn_usage:int
    device_changes_last_30_days:int
    suspicious_contacts_count:int
    avg_call_duration_min:float
    microphone_Activation_events:int
    camera_activation_events:int
    account_login_failures:int
    location_change_frequency:float
    military_zone_proximity_score:float
    year:int
    month:int
    day:int
    hour:int
    minute:int
    second:int

# Home route
@app.get("/")
def home():
    return {"message": "Surveillance Model API is running"}

# Prediction route
@app.post("/predict")
def predict(data: InputData):
    # Convert input to array
    input_array = np.array([
         person_id:int
    call_count_with_suspects:int
    night_movements_hours:float
    gps_zone_visits:int
    encrypted_app_usage_hours:float
    social_media_threat_score:float
    vpn_usage:int
    device_changes_last_30_days:int
    suspicious_contacts_count:int
    avg_call_duration_min:float
    microphone_Activation_events:int
    camera_activation_events:int
    account_login_failures:int
    location_change_frequency:float
    military_zone_proximity_score:float
    year:int
    month:int
    day:int
    hour:int
    minute:int
    second:int
    ]).reshape(1, -1)

    # Make prediction
    prediction = model.predict(input_array)

    return {
        "prediction": prediction.tolist()
    }
```
